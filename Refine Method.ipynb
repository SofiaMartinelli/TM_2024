```python
# Función para la creación de bloques

from nltk.tokenize import sent_tokenize

# Cuenta el número de palabras en un texto.
def count_words(text):
    return len(text.split())

# Crea bloques de texto basado en oraciones, asegurando que cada bloque no exceda max_length palabras.
def create_blocks(text, max_length=512):
    # Dividir el texto en oraciones
    sentences = sent_tokenize(text, language='spanish')
    blocks = []
    current_block = []
    current_length = 0

    for sentence in sentences:
        sentence_length = count_words(sentence)

        # Si agregar esta oración excede el límite de palabras, crear un nuevo bloque
        if current_length + sentence_length > max_length:
            if current_block:
                blocks.append(" ".join(current_block))
                current_block = []
                current_length = 0

        # Agregar la oración al bloque actual
        current_block.append(sentence)
        current_length += sentence_length

    # Agregar el último bloque si no está vacío
    if current_block:
        blocks.append(" ".join(current_block))

    return blocks
```

```python
# Función para generar resúmenes utilizando el modelo RoBERTa.
def generate_summary(text, prompt, max_length=150, num_beams=5):
    # Combinar el prompt con el texto.
    full_prompt = f"{prompt}\n\n{text}"

    # Tokenizar el texto.
    inputs = tokenizer([full_prompt], padding="max_length", truncation=True, max_length=512, return_tensors="pt")
    input_ids = inputs.input_ids.to(device)
    attention_mask = inputs.attention_mask.to(device)

    # Generar el resumen utilizando el modelo con parámetros ajustados.
    output = model.generate(
        input_ids,
        attention_mask=attention_mask,
        max_length=max_length,  # Longitud máxima configurable
        min_length=100,
        length_penalty=2.0,
        num_beams=num_beams,     # Número de beams configurable
        do_sample=True,          # Activa el muestreo para mayor variabilidad
        early_stopping=True      # Detiene la generación si se obtiene un resultado adecuado
    )

    # Decodificar el resumen generado y eliminar los tokens especiales.
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Función para aplicar el método "Refine" y generar un resumen iterativo.
def refine_summary(text_blocks, prompt):
    # Empezar con el resumen del primer bloque.
    i = 1
    current_summary = generate_summary(text_blocks[0], prompt)
    print(f"Resumen {i}:\n{current_summary}")
    i += 1

    # Iterar sobre los siguientes bloques de texto.
    for block in text_blocks[1:]:
        # Combinar el resumen actual con el siguiente bloque.
        combined_text = current_summary + "\n" + block

        print(f"Texto combinado:\n{combined_text}")

        # Generar un nuevo resumen de la combinación.
        current_summary = generate_summary(combined_text, prompt)
        print(f"Resumen {i}:\n{current_summary}")
        i += 1

    # Al final, el resumen contendrá el resultado refinado.
    return current_summary
```
